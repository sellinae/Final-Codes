### This code sould be used on to any selected data.
### It will perform data augmentation randomly according to the rules specefied.

```python
import numpy as np
import matplotlib.pyplot as plt
import soundfile as sf
import random

def proba(p): # generate probability
    if random.randint(0,100) <= p:
        return True
    else:
        return False

def save_data(data, samplerate, export_path): # save the data 
    sf.write(export_path, data, samplerate)
    return

def load_data(file_path): # read the data
    data, samplerate = sf.read(file_path)
    # print("data length:", len(data))
    # print("sample rate:", samplerate)
    return data, samplerate

def plot_time_series(data):
    plt.figure(figsize=(14, 8), facecolor='white')
    plt.title('Raw wave ')
    plt.ylabel('Amplitude')
    plt.plot(np.linspace(0, 1, len(data)), data)
    plt.show()
    return

def manipulate(data, noise_factor): # add noise to sound file
    noise = np.random.randn(len(data))
    augmented_data = data + noise_factor * noise
    # Cast back to same data type
    augmented_data = augmented_data.astype(type(data[0]))
    return augmented_data

def shift_sound(data, shift_factor): # shift sound file
    shift_amount = len(data) * shift_factor
    shift_amount = int(shift_amount)
    data_shift = np.roll(data, shift_amount)
    return data_shift
```

```python
import os, shutil
from tqdm import tqdm

def augment_randomly(full_path, export_path):
    for filename in tqdm(os.listdir(full_path)):
        if filename.endswith(".wav") or filename.endswith(".WAV"): # Check if the file ends with .wav
            #print("Full path:", full_path + filename)
            save_file_name = filename[:-4]
            
            # setting the augmentation threshold (yes/no):
            aug_threshold = 50
            
            # setting shift threshold:
            shift_threshold = 50
            
            # setting noise threshold:
            manipulate_threshold = 50
            
            # flags
            shift_flag = False
            manipulate_flag = False
            
            data, samplerate = load_data(full_path + filename)
            if proba(aug_threshold):
                if proba(shift_threshold):
                    # shift_factor = 0.7
                    shift_factor = random.random()
                    # shift_factor = random.uniform(0.2, 0.7)
                    data = shift_sound(data, shift_factor)
                    shift_flag = True
                if proba(manipulate_threshold):
                    noise_factor = random.uniform(0.0001, 0.002)
                    # noise_factor = 0.0001
                    data = manipulate(data, noise_factor)
                    manipulate_flag = True
                # if shift_flag == True and manipulate_flag == True:
                #     save_data(data, samplerate, export_path + save_file_name + '_shift_aug.wav')
                #     #print("Saving...", export_path + save_file_name + '_shift_aug.wav')
                # elif shift_flag == True:
                if shift_flag == True:
                    save_data(data, samplerate, export_path + save_file_name + '_shift.wav')
                    #print("Saving...", export_path + save_file_name + '_shift.wav')
                elif manipulate_flag == True:
                    save_data(data, samplerate, export_path + save_file_name + '_aug.wav')
                    #print("Saving...", export_path + save_file_name + '_aug.wav')
            # else:
            #     print("Augmentation probability threshold not met, skipping file.")
            shift_flag = False
            manipulate_flag = False
    return
```

```python
import os
import shutil
from tqdm import tqdm

full_path = r'split_data_increased/HT/' 
export_path = r'split_data_aug/HT/'

if not os.path.exists(export_path):
    os.makedirs(export_path)

# # fetch all files
# for file_name in tqdm(os.listdir(full_path)):
#     # construct full file path
#     source = full_path + file_name
#     destination = export_path + file_name
#     # copy only files
#     if os.path.isfile(source):
#         shutil.copy(source, destination)

print('all files were copied successfully')
augment_randomly(full_path, export_path)
```
